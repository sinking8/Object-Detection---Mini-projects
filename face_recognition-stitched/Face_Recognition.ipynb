```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
from PIL import Image
```

```python
#importing images
image1 = Image.open('./images/face1.jpg')
image2 = Image.open('./images/face2.jpg')
```

```python
def merge_imgs(image1,image2):
    image1_size = image1.size
    image2 = image2.resize(image1_size)
    merged_image = Image.new('RGB',(2*image1_size[0], image1_size[1]), (250,250,250))
    merged_image.paste(image1,(0,0))
    merged_image.paste(image2,(image1_size[0],0))
    return merged_image    
```

```python
#Cascade Classifier
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_faces(img):

    gray = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    faces  = face_cascade.detectMultiScale(gray,1.1,4)

    font = cv2.FONT_HERSHEY_SIMPLEX 

    for(x,y,w,h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

    cv2.imwrite('detected.png',img)
```

```python
#Displaying First Image
plt.title("Image1")
plt.imshow(image1)
```

```python
#Displaying Second Image
plt.title("Image2")
plt.imshow(image2)
```

```python
#Merged Image
new_img = merge_imgs(image2,image1)
```

```python
#Displaying Stitched Image
plt.title("Stitched Image")
plt.imshow(new_img)
```

```python
img = np.array(new_img)
detect_faces(img)
```

```python
#Displaying Detected Image
img = cv2.imread('detected.png')
plt.title("Detected Image")
plt.imshow(img)
```
