```python
import cv2
import numpy as np 
from PIL import Image
```

```python
img = cv2.imread(os.getcwd()+'/data/Task_3_Pipe_Counting.jpeg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_blurred = cv2.blur(gray, (3, 3)) 
  
detected_circles = cv2.HoughCircles(gray_blurred,  
                   cv2.HOUGH_GRADIENT, 1, 20, param1 = 50, 
               param2 = 30, minRadius = 10, maxRadius = 30) 
  
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
if detected_circles is not None: 
  
    detected_circles = np.uint16(np.around(detected_circles)) 
    print("Total Number of Pipes %d"%(detected_circles.shape[1]))
    
    for pt in detected_circles[0, :]: 
        a, b, r = pt[0], pt[1], pt[2] 
        cv2.circle(img, (a, b), r, (0, 255, 0), 2)  
    cv2.imwrite('Detected.png', img);    
```

```python
final_img = cv2.imread("Detected.png")
plt.imshow(final_img)
```
